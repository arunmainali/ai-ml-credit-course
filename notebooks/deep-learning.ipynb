# Deep Learning (implemented through neural networks)

## Types of Deep Learning
- ANN/DNN: User for all kinds of data
- RNN/LSTM: Used for NLP tasks, linguistics data
- CNN (Computer Vision): Used for visuals data (images/videos)

```python
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = data.load_data()
```

```python
train_images.shape, test_images.shape
train_labels.shape, test_labels.shape
```

```python
class_name = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

\# of classes = 10

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28, 28)),
    keras.layers.Dense(128, activation = 'relu'),
    keras.layers.Dense(10, activation = 'softmax') # softmax: gives prbability distribution
])
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs = 10)
```

```python
prediction = model.predict(test_images)
```

```python
for i in range(10):
    plt.imshow(test_images[i])
    plt.xlabel('Actual: ' + class_name[test_labels[i]])
    plt.title('Prediction: ' + class_name[np.argmax(prediction[i])])
    plt.show()
```
