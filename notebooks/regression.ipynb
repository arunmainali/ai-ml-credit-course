```python
# Regression

df = pd.read_csv('../data/student_performent_prediction/mat2.csv')

df

df.shape

df.columns

# featuer selection
data = df[['G1', 'G2', 'G3', 'studytime', 'failures', 'absences']]

X = data.drop(columns = 'G3')
y = data['G3']

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model

model.fit(X.values, y.values)
model.predict([[20, 15, 12, 15, 18]])

## Test Accuracy
### Some data for prediction and some for testing

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train.values, y_train.values)
prediction = model.predict(X_test.values)

## errors and accuracy
### exactly milyo ki milena : accuracy

#### jhandai mulyo ki milena : mean sq error, mean absoulte errror, r2 score

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(prediction, y_test)
mse

model.coef_

model.intercept_

import numpy as np
X_test = np.array(X_test)
y_test = np.array(y_test)

for i in range(len(prediction)):
    print(X_test[i], y_test[i], prediction[i])
```
